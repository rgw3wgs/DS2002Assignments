```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('top_100_movies_by_genres.csv')
```

```python
print(df.head())
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(df['RatingTomatometer'], df['No. of Reviews'], color='blue', alpha=0.5)
plt.title('Rating vs. No. of Reviews')
plt.xlabel('Rating')
plt.ylabel('No. of Reviews')

num_ticks = 10
plt.xticks(df['RatingTomatometer'].unique()[::len(df['RatingTomatometer'].unique())//num_ticks])

plt.xticks(rotation=45, ha='right')

plt.grid(True)
plt.show()
```

```python
Ttop_10_reviews = df.nlargest(10, 'No. of Reviews')
plt.figure(figsize=(10,6))
plt.bar(top_10_reviews['Title'], top_10_reviews['No. of Reviews'], color='skyblue')
plt.title('Top 10 Movies by Number of Reviews')
plt.xlabel('Movie Title')
plt.ylabel('Number of Reviews')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
genre_counts = df['Genre'].value_counts()
plt.figure(figsize=(10, 8))
plt.pie(genre_counts, labels=genre_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Movie Genres')
plt.axis('equal')
plt.show()
```
